```python
from google.colab import userdata
api_key:str=userdata.get('GOOGLE_API_KEY_2')
if(api_key):
  print("api key is connect")
else:print("api key is not found")
```

```python
from google import genai
from google.genai import Client
client:Client=genai.Client(api_key=api_key)
model="gemini-2.0-flash-exp"
```

```python
video="/content/video#1.mp4"
```

```python
import time
def upload_video(video):
 video_file=client.files.upload(path=video)
 while video_file.state=="Proccessing":
   print("waiting for a video to process")
   time.sleep(15)
   video_file=client.files.get(name=video_file.name or "")
 if video_file.state=="Failed":
  raise ValueError(video_file.state)
 print(f'Video proccesssing complete:'+(video_file.uri or ""))
 return video_file
my_video=upload_video(video)
```

```python
from google.genai.types import Content, Part
prompt="har seen ka liye is video ma ek caption generate kro aur is ma bola gya har seen quotation marks ma likho har caption ko video ka time code ke sath ek object ma rakho"
VIDEO=my_video
response=client.models.generate_content(
    model=model,
    contents=[
        Content(
            role="user",
            parts=[
                Part.from_uri(
                    file_uri=my_video.uri or "",
                    mime_type=my_video.mime_type or ""
                ),
            ]
        ),
        prompt,
    ]
)
print(response.text)
```
